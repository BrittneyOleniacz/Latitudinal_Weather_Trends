### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# Configure gmaps
gmaps.configure(api_key = g_key)
# Heatmap of humidity
locations = DF_city_weather_data[["Latitude", "Longitude"]]
humidity = DF_city_weather_data["Humidity"]

fig = gmaps.figure(center=(40.6937, -75.1902), zoom_level=2)
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=300, point_radius=5)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
filtered = DF_city_weather_data.loc[(DF_city_weather_data["Max. Temp."] >= 55) & (DF_city_weather_data["Max. Temp."] <= 65) & (DF_city_weather_data["Humidity"] > 80) 
                                   & (DF_city_weather_data["Wind"] < 10) & (DF_city_weather_data["Cloudiness"] >= 70), :]
all_filtered = filtered.reset_index()
all_filtered
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = all_filtered
hotel_df["Hotel Name"] = ""

# Set parameters to search for a hotel
params = {
    "radius": 5000,
    "types": "lodging",
    "key": g_key,
     }
```

```python
# Iterate through 
for index, row in hotel_df.iterrows():
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    params["location"] = f"{latitude}, {longitude}" 
    # Use the search term: "Hotel" and our lat/lng
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    # make request and print url
    name_address = requests.get(base_url, params=params)
    # convert to json
    name_address = name_address.json()
    # Grab the first hotel from the results and store the name
    try:
        hotel_df.loc[index, "Hotel Name"] = name_address["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
```

```python
# Add marker layer ontop of heat map
markers = gmaps.marker_layer(locations, info_box_content=hotel_info)
fig.add_layer(markers)
# Display figure"
fig
```
