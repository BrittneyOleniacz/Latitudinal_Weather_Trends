```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import scipy.stats as st
from scipy.stats import linregress
from datetime import datetime as dt

# Import API key
from API_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
 # List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + weather_api_key 
# List of city data
city_data = []
city = "Lawrence"

for city in cities:
    # Create endpoint URL with each city
    city_url = url + "&q=" + city
    # Run an API request for each of the cities
    try:
        # Parse the JSON and retrieve data
        city_weather = requests.get(city_url).json()
        city_dict = {
        "name" : city,
        "lat" : city_weather["coord"]["lat"],
        "long" : city_weather["coord"]["lon"],
        "max_temp" : city_weather["main"]["temp_max"],
        "humidity" : city_weather["main"]["humidity"],
        "clouds" : city_weather["clouds"]["all"],
        "wind_speed" : city_weather["wind"]["speed"],
        "Country" : city_weather["sys"]["country"],
        "date" : city_weather["dt"]}    
        city_data.append(city_dict)
    except:
        print(f"data not found")       
city_data  
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
len(city_data)
```

```python
#Create, format, arrange Dataframe
city_weather_df = pd.DataFrame(city_data).rename(columns = {
"name" : "City Name",
"country" : "Country",
"lat" : "Latitude",
"long" : "Longitude",
"max_temp" : "Max. Temp.",
"humidity": "Humidity",
"clouds": "Cloudiness",
"wind_speed" : "Wind",
"date": "Linux_timestamp"}, inplace = False)
city_weather_df["City Name"] = city_weather_df["City Name"].str.capitalize()
city_weather_df = city_weather_df[["City Name", "Country", "Latitude", "Longitude", "Max. Temp.", "Humidity", "Cloudiness", "Wind", "Linux_timestamp"]]

#Convert Linux timestamp to date
conversion = pd.to_datetime(city_weather_df["Linux_timestamp"], unit='s')
city_weather_df["Date_time"] = pd.Series(conversion, index=city_weather_df.index) 
city_weather_df['Time'],city_weather_df['Date']= city_weather_df['Date_time'].apply(lambda x:x.time()), city_weather_df['Date_time'].apply(lambda x:x.date())
del city_weather_df["Date_time"], city_weather_df["Linux_timestamp"], city_weather_df["Time"]

city_weather_df.to_csv("city_weather.csv")
city_weather_df
```

## Linear Regression

Your second requirement is to run linear regression on each relationship, only this time separating them into Northern Hemisphere (greater than or equal to 0 degrees latitude) and Southern Hemisphere (less than 0 degrees latitude):

* Northern Hemisphere - Temperature (F) vs. Latitude
* Southern Hemisphere - Temperature (F) vs. Latitude
* Northern Hemisphere - Humidity (%) vs. Latitude
* Southern Hemisphere - Humidity (%) vs. Latitude
* Northern Hemisphere - Cloudiness (%) vs. Latitude
* Southern Hemisphere - Cloudiness (%) vs. Latitude
* Northern Hemisphere - Wind Speed (mph) vs. Latitude
* Southern Hemisphere - Wind Speed (mph) vs. Latitude

After each pair of plots explain what the linear regression is modeling such as any relationships you notice and any other analysis you may have.

```python
north_hemisphere = city_weather_df.loc[(city_weather_df["Latitude"]) >= 0]
south_hemisphere = city_weather_df.loc[(city_weather_df["Latitude"]) <= 0]
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
slope, intercept, r_value, p_value, std_err = st.linregress(north_hemisphere["Latitude"], north_hemisphere["Max. Temp."])
print(f"The r-value is: {round(r_value, 2)}")

fig1, ax1 = plt.subplots(figsize = (15, 10))
model_fit = north_hemisphere["Latitude"] * slope + intercept
plt.scatter(north_hemisphere["Latitude"], north_hemisphere["Max. Temp."], s = 175, color = "firebrick")
plt.plot(north_hemisphere["Latitude"], model_fit, linewidth =4 , color = "dimgrey")
plt.title("Northern Hemisphere - Temperature (F) vs. Latitude", fontsize = 20)
plt.xlabel("Latitude", fontsize = 14)
plt.ylabel("Temperature (F)", fontsize = 14)
equation = "y = " + str(round(slope,2)) + "x" " + " + str(round(intercept,2))
plt.annotate(equation, (4, 96), fontsize = 14, fontweight = "bold", color = "dimgrey")
plt.savefig("Northern_Hemisphere - Latitude_vs._Temperature.png")

plt.show()
```

There's an inverse relationship between latitude and temperature; thus, the higher the latitude, the lower the temperature. Conversely, the lower the latitude, the higher the temperature. 

####  Southern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
slope, intercept, r_value, p_value, std_err = st.linregress(south_hemisphere["Latitude"], south_hemisphere["Max. Temp."])
print(f"The r-value is: {round(r_value, 2)}")

fig1, ax1 = plt.subplots(figsize = (15, 10))
model_fit = south_hemisphere["Latitude"] * slope + intercept
plt.scatter(south_hemisphere["Latitude"], south_hemisphere["Max. Temp."], s = 175, color = "firebrick")
plt.plot(south_hemisphere["Latitude"], model_fit, linewidth = 4, color = "dimgrey")
plt.title("Southern Hemisphere - Temperature (F) vs. Latitude", fontsize = 20)
plt.xlabel("Latitude", fontsize = 14)
plt.ylabel("Temperature (F)", fontsize = 14)
equation = "y = " + str(round(slope,2)) + "x" " + " + str(round(intercept,2))
plt.annotate(equation, (-50, 61), fontsize = 14, color = "dimgrey", fontweight = "bold")
plt.savefig("Southern_Hemisphere - Latitude_vs._Temperature.png")
plt.show()
```

The latitude and temperature of the southern hemisphere are directly correlated because the lower the latitude, the lower the temperature.

Wind speed is the highest at the greater latitudes. These regions are mostly tundras and glacial deserts (i.e., Siberia) because of the lack of direct sun. So, there is a lack of interference and competing forces, and many sources of wind. The air is denser in colder areas and sinks to the Earth's surface, thus adding to the surface winds along an orthogonal plane. Additionally, the Earth's rotation creates wind currents forced laterally from the equator to the poles. With a smaller circumference at the poles, the centripetal forces make for more compact air cells, intensifying the wind speeds.
